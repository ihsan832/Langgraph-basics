```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs:int
    balls:int
    four:int
    sixes:int
    sr:float
    bpb:float
    boundery_percent:float
    summary:str
```

```python
def calculate_sr(state:BatsmanState):
    sr = (state['runs']/state['balls'])*100
    return {'sr':sr}
```

```python
def calculate_bpb(state:BatsmanState):
    bpb = state['balls'] / (state['four'] + state['sixes'])

    return {'bpb':bpb}
```

```python
def boundery_percent(state:BatsmanState):
    boundery_percent = (((state['four'] * 4) + (state['sixes'] * 6)) /state['runs']) * 100
    return {'boundery_percent':boundery_percent}
```

```python
def summary(state:BatsmanState):
    summary = f"""
Strike Rate - {state['sr']} 
Balls per boundery - {state['bpb']} 
Boundery Percent - {state['boundery_percent']}
"""
    
    return {'summary':summary}
```

```python
graph = StateGraph(BatsmanState)
graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_bpb',calculate_bpb)
graph.add_node('boundery_percent',boundery_percent)
graph.add_node('summary',summary)
```

```python
graph.add_edge(START,'calculate_sr')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'boundery_percent')
```

```python
graph.add_edge('calculate_sr','summary')
graph.add_edge('calculate_bpb','summary')
graph.add_edge('boundery_percent','summary')
```

```python
graph.add_edge('summary',END)
```

```python
workflow = graph.compile()
```

```python
initial_state = {
    'runs':100,
    'balls':50,
    'four':5,
    'sixes':5,
}
```

```python
final_state = workflow.invoke(initial_state)
```
